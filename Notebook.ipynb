# KMeans Clustering on the Data

```python
from sklearn.cluster import KMeans
import sklearn.metrics as metrics
```

```python
col=['cook', 'eating_out',
       'fruit_day', 'income']
```

```python
Data = Data[col]
sos = []
silh = []
l=[]
for i in range(2,12):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 500, n_init = 10, random_state = 2)
    lables = kmeans.fit_predict(Data)
    n=[sum(lables == j) for  j in range(1,i+1)]
    l.append(n)
    sos.append(kmeans.inertia_)
    silh.append(metrics.silhouette_score(Data,lables,metric='euclidean',sample_size=1000, random_state=20))

plt.figure(figsize=(15,10))
plt.plot(range(2,12),sos)
```

```python
k = [i*100 for i in np.diff(sos,2)].index(min([i*100 for i 
     in np.diff(sos,2)]))
k
```

```python
plt.plot(range(2,12),silh)
```

```python
venues = result['response']['groups'][0]['items']
nearby_venues = json_normalize(venues)
nearby_venues
```

```python
rest=[]
other=[]
for lat,long in zip(nearby_venues['venue.location.lat'],nearby_venues['venue.location.lng']):
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
      client_id, 
      client_secret, 
      VERSION, 
      lat,long,
      1000, 
      100)
    res = requests.get(url).json()
    venue = res['response']['groups'][0]['items']
    nearby_venue = json_normalize(venue)
    df=nearby_venue['venue.categories']

    g=[]
    for i in range(0,df.size):
        g.append(df[i][0]['icon']['prefix'].find('food'))
    co=0
    for i in g:
        if i>1:
            co+=1
    rest.append(co)
    other.append(len(g)-co)

nearby_venues['restaurant']=rest
nearby_venues['others']=other
nearby_venues
```

```python
colu = ['venue.location.lat','venue.location.lng','venue.location.formattedAddress','restaurant','others']
```

```python
nearby_venues = nearby_venues[colu]
nearby_venues = nearby_venues.rename(columns = {'venue.location.lat' : 'lat',
                                               'venue.location.lng' : 'lng',
                                               'venue.location.formattedAddress' : 'address'})
nearby_venues
```

```python
data = nearby_venues[['lat','lng','restaurant','others']]

kmeans = KMeans(n_clusters = 6, init = 'k-means++', max_iter = 500, n_init = 10, random_state = 2)
l=kmeans.fit_predict(data)
```

```python
nearby_venues['cluster'] = l
nearby_venues
```
